```python
!pip install datasets --quiet
from datasets import load_dataset

dataset = load_dataset("cnn_dailymail", "3.0.0", split='train', streaming=True)
first10 = dataset.take(10)
pd.DataFrame(first10)
```

## Generate our own Summaries

1. Write code to create summaries for the ten Daily Mail reviews and add them to the dataframe, along with the original articles

```python
df = pd.DataFrame(first10)
summaries = []
for i in range(10):
    summaries.append(summarizer(df['article'][i][0:2500], min_length=100, max_length=200)[0]['summary_text'])
df['summary'] = summaries
df
```

2. Display all ten summaries (maybe use a for loop and a print statement)


```python
for i in range(10):
  print(f'Article {i+1} Summary:')
  print(summaries[i])
```

## Evaluate our Summaries using ROUGE-2

Here's an example of how to use ROUGE.

```python
!pip install rouge --quiet

from rouge import Rouge

model_out = "he began by starting a five person war cabinet and included chamberlain as lord president of the council"
reference = "he began his premiership by forming a five-man war cabinet which included chamberlain as lord president of the council"

rouge = Rouge()
rouge.get_scores(model_out, reference)
```

## Calculate the Average ROUGE-2 score across the first 10 generated summaries

3. Calculate the ROUGE-2 score for each of the ten summary/reference pairs, using the provided summary as a reference and add the score (just the 'r' part of the ROUGE-2 part of the resulting dictionary) to the dataframe


```python
rouge_2_scores = []
rouge_2_r = []

for i in range(10):
    model_out = df['summary'][i]
    reference = df['highlights'][i]

    rouge = Rouge()
    score = rouge.get_scores(model_out, reference)
    rouge_2_scores.append(score[0]['rouge-2'])
    rouge_2_r.append(score[0]['rouge-2']['r'])

# Print ROUGE-2 score for each of the ten summary/reference pairs
for i in range(10):
  print(f'Article {i+1} ROUGE-2 score:')
  print(rouge_2_scores[i])
```

```python
# Adding the 'r' score to the dataframe
df['rouge_2_score'] = rouge_2_r
df
```

4. Calculate the average of the ten ROUGE-2 'r' scores


```python
np.mean(df['rouge_2_score'])
```

```python
from evaluate import load
bertscore = load("bertscore")
summary = df['summary']
references = df['highlights']
results = bertscore.compute(predictions=summary, references=references, lang="en", model_type="distilbert-base-uncased")

# Print BERT score for the ten summary/reference pairs
print(f'BERT Scores:')
print(results)
```

```python
# Adding the 'F1' score to the dataframe
df['bert_F1_score'] = results['f1']
df
```

Calculate the average of the ten BERT 'F1' scores
